```python
import math
import pyaudio
import struct
import time

# Parameters
volume = 0.5       # volume level (0.0 to 1.0)
fs = 44100         # sampling rate, Hz, must be an integer
duration = 1.0     # duration of the signal in seconds
frequency = 440.0  # frequency of the sine wave in Hz (A4 note)




# Generate the sine wave
num_samples = int(fs * duration)

sine_wave = [
    volume * math.sin(x * ((2 * math.pi) * frequency) / fs)
    for x in range(num_samples)
]




# Convert to bytes
audio_data = b''.join(
    [struct.pack('f', sample) for sample in sine_wave]
)

# Initialize PyAudio
p = pyaudio.PyAudio()

# Open a stream with the appropriate settings
stream = p.open(format=pyaudio.paFloat32,
                channels=1,
                rate=fs,
                output=True)

# Play the audio data

for _ in range(2):
    time.sleep(1)
    stream.write(audio_data)

# Close the stream
stream.stop_stream()
stream.close()

# Terminate the Py
# Audio session
p.terminate()
```
